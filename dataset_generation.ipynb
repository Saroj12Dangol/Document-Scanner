```python
import os
import numpy as np
import cv2
from glob import glob
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
```

```python
path = '/home/saroj/Downloads/annotation_3_json.json'
```

```python
f = open(path, "r")
data = json.load(f)
data
```

```python
for key, value in tqdm(data.items()):
    """ Getting the filename """
    filename = value["filename"]
    regions = value["regions"]

    image = cv2.imread(f"/home/saroj/Desktop/Dataset/Images/{filename}", cv2.IMREAD_GRAYSCALE)
    H, W = image.shape
    mask = np.zeros((H, W))

    for region in regions:
        # print(region)
        x = int(region["shape_attributes"]["x"])
        y = int(region["shape_attributes"]["y"])
        width = int(region["shape_attributes"]["width"])
        height = int(region["shape_attributes"]["height"])
        print(x, y, x+width, y+height)

        mask = cv2.rectangle(mask, (x, y), (x+width, y+height), (255, 255, 255), -1)
    
    cv2.imwrite(f'/home/saroj/Desktop/Dataset/mask/{filename}', mask)
# plt.figure(figsize=(20,20))
# plt.imshow(mask, cmap='gray')
```
